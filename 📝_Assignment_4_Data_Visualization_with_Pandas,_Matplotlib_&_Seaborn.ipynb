1. Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

2. Create a DataFrame

```python
Data = {
    'product':['A', 'B', 'C', 'D'],
    'Sales':[25000, 30000, 15000, 20000],
    'Profit':[4000, 6000, 2000, 3000]
}
df = pd.DataFrame(Data)
print(df)
```

3. Plot a Bar Chart (Using Pandas)

```python
df.plot(kind='bar',x='product',y='Sales',title='Sales per product')
plt.ylabel('Sales Amount')
plt.grid(True)
plt.show()
```

4. Line Chart (Using Pandas)


```python
df.plot(kind='line', x='product', y='Profit', title='Profit Trend')
plt.ylabel('Profit')
plt.grid(True)
plt.show()
```

5. Seaborn Barplot


```python
sns.barplot(x='product', y='Sales', data=df)
plt.title('Sales per product(Seaborn)')
plt.ylabel('Sales Amount')
plt.grid(True)
plt.show()
```

6. Boxplot (To check distribution of Sales and Profit)

```python
sns.boxplot(data=df[['Sales', 'Profit']])
plt.title('Distribution of Sales and Profit')
plt.grid(True)
plt.show()
```

7. Pie Chart (Using Pandas)



```python
df.set_index('product')['Sales'].plot(kind='pie', autopct='%1.1f%%', title=' Sales Distribution')
plt.ylabel('')
plt.show()
```
